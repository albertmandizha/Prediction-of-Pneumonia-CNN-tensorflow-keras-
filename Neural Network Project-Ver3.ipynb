### Notes Observed From the dataset

Imbalanced Data set with pneumonia class being larger than the normal class. The following challenges can be encountered.

Bias towards majority class: Neural networks trained on imbalanced datasets tend to be biased towards the majority class, resulting in poor performance on the minority class. This is because the model is optimized to minimize the overall loss function, which can lead to the majority class dominating the decision-making process.

Difficulty in capturing minority class: The imbalance in the dataset can make it difficult for the neural network to learn the features of the minority class. This can lead to poor accuracy, sensitivity, and recall for the minority class.

Inflated accuracy: Accuracy alone is not an appropriate metric for evaluating the performance of imbalanced datasets since the model can achieve high accuracy by simply predicting the majority class. Therefore, evaluation metrics such as precision, recall, F1 score, and AUC should be used instead.

Generalization issues: Imbalanced datasets can lead to overfitting since the model can learn to classify the training set accurately while ignoring the minority class. This can result in poor generalization performance on new data.

To overcome these challenges, techniques such as data resampling (e.g., oversampling, undersampling), cost-sensitive learning, and ensemble methods can be used to balance the dataset and improve the performance of neural networks on imbalanced datasets.

### Creating a funtion get_training_data that takes a directory path as input and returns a NumPy array

This code defines a function named get_training_data that takes a directory path as input and returns a NumPy array containing the training data in the form of image arrays and corresponding labels.

The function starts by defining a list of labels that correspond to the different classes in the dataset (in this case, 'PNEUMONIA' and 'NORMAL') and an image size of 200x200 pixels.

It then creates an empty list called data to store the training data. The function loops over each label in the list of labels and gets the corresponding path for the label by joining the input data directory path with the label.

The function then obtains the index of the label in the list of labels and assigns it to a variable called class_num.

Inside the inner loop, the function reads each image file in the directory using the OpenCV cv2.imread function and converts it to grayscale using the cv2.IMREAD_GRAYSCALE flag.

The function then resizes the image array to the specified image size using the OpenCV cv2.resize function and adds the resized image array and its corresponding label (as a number) to the data list.

If an exception occurs during the image reading or resizing process, the function prints the exception message and continues to the next image.

Finally, the function returns the training data as a NumPy array of image arrays and corresponding class labels.

```python
labels = ['PNEUMONIA' , 'NORMAL']
img_size = 200


def get_t_data(data_dir):
    data = []
    for label in labels:
        path = os.path.join(data_dir, label)
        class_num = labels.index(label)
        for img in os.listdir(path):
            try:
                img_arr = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
                resized_arr = cv2.resize(img_arr, (img_size, img_size))
                data.append([resized_arr, class_num])
            except Exception as e:
                print(e)
    return np.array(data)
```

### Notes from the Image Size
The image size can have a significant effect on the neural network's performance, both in terms of accuracy and training time. Here are some ways in which the image size can affect the neural network:

Accuracy: A larger image size can potentially provide more information for the neural network to learn from, resulting in higher accuracy. However, this is not always the case, as a larger image size can also increase the complexity of the network, making it more difficult to train and causing it to overfit to the training data.

Training time: A larger image size can also increase the training time of the neural network, as more pixels need to be processed for each image. This can be especially problematic when dealing with large datasets, as training the network on a large number of images with large dimensions can take a long time and require significant computational resources.

Model size: The image size can also affect the size of the neural network model. A larger image size will result in a larger number of input neurons, which in turn can increase the number of parameters in the model. This can make the model more complex and difficult to train, especially if the dataset is small.

In general, choosing the optimal image size for a neural network requires a trade-off between accuracy, training time, and model size. It is important to experiment with different image sizes and monitor the performance of the network to find the optimal image size for a particular problem

## 5.0 Creation of the CNN

```python
#First Training Time with epochs of 2
```

```python
# Creation of CNN
CNN = tf.keras.models.Sequential ()
# Adding Convolution layer
CNN.add (tf.keras.layers.Conv2D (filters = 32, kernel_size = 3, activation = "relu", input_shape = [256,256,3]))
# Adding Pooling layer
CNN.add (tf.keras.layers.MaxPool2D (pool_size = 2 , strides = 2))
# Adding Flattening layer
CNN.add (tf.keras.layers.Flatten ())
# Adding Full connection layer
CNN.add (tf.keras.layers.Dense (units = 512 , activation ="relu"))
# Adding Output Layer
CNN.add (tf.keras.layers.Dense (units = 2 , activation ="sigmoid"))
CNN.summary ()
CNN.save("cnn_model.bin")
```

This code creates a Convolutional Neural Network (CNN) model using the Keras library in TensorFlow. Here is what each line of the code does:

CNN = tf.keras.models.Sequential(): This creates a new Sequential model in Keras, which allows us to build a neural network layer by layer.

CNN.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu", input_shape=[256,256,3])): This adds a Conv2D layer to the model with 32 filters, a kernel size of 3x3, ReLU activation function, and an input shape of [256,256,3]. This layer performs convolution operation on the input image to extract features.

CNN.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2)): This adds a MaxPooling2D layer to the model with a pool size of 2x2 and stride of 2. This layer reduces the spatial size of the output volume and helps to prevent overfitting.

CNN.add(tf.keras.layers.Flatten()): This adds a Flatten layer to the model, which flattens the output of the previous layer into a 1-dimensional vector. This layer prepares the data for the fully connected layers.

CNN.add(tf.keras.layers.Dense(units=512, activation="relu")): This adds a Dense layer with 512 units and a ReLU activation function. This layer is a fully connected layer, which learns the patterns in the data.

CNN.add(tf.keras.layers.Dense(units=2, activation="sigmoid")): This adds another Dense layer with 2 units and a sigmoid activation function. This is the output layer, which predicts the probability of the input image belonging to each of the two classes (pneumonia and normal).

CNN.summary(): This prints a summary of the model, including the layer type, output shape, and number of parameters.

CNN.save("cnn_model.bin"): This saves the trained model to a file named "cnn_model.bin" in the current working directory

## Training the CNN

#### Code comments

What does the Trainig code do?

This code defines a convolutional neural network (CNN) model, compiles it, and trains it on the given training data (Train) and validation data (Validation) for 2 epochs.

The CNN.compile() function compiles the CNN model with an optimizer, a loss function, and metrics. The optimizer determines how the model updates its parameters during training, the loss function measures how well the model is performing during training, and the metrics evaluate the performance of the model after each epoch. In this case, the optimizer is set to "adam", which is a popular optimizer used in deep learning. The loss function is "binary_crossentropy", which is a commonly used loss function for binary classification tasks, and the metrics are set to "accuracy", which is a measure of how well the model predicts the correct class labels.

The CNN.fit() function trains the compiled model using the training data (Train) and validates it using the validation data (Validation) for the specified number of epochs (in this case, 2). During training, the model is adjusted to minimize the loss function by updating its weights and biases. The validation data is used to evaluate the model's performance after each epoch, and the metrics are computed and recorded for both the training and validation sets.
WHY THE EPOCHS OF 2?

The number of epochs determines how many times the model will be trained on the entire dataset. Each epoch consists of a forward pass, where the model makes predictions on the input data, and a backward pass, where the model's parameters are updated based on the error between the predicted output and the actual output.

The choice of the number of epochs depends on various factors such as the size of the dataset, the complexity of the model, the learning rate, and the training time. In general, increasing the number of epochs can improve the model's performance, but training for too many epochs can cause overfitting, where the model starts to fit the noise in the training data instead of the underlying patterns.

In this case, the model is trained for 2 epochs, which might be sufficient to get a sense of how the model is performing on the given data. However, the optimal number of epochs can only be determined through experimentation, and it might require more than two epochs to achieve the desired performance.

```python
### Training the CNN
CNN.compile (optimizer = "adam" , loss ="binary_crossentropy" , metrics = ["accuracy"])
CNN.fit (x = Train , validation_data = Validation , epochs = 2)
```
